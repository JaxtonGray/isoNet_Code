IsoP Data Request
=================
Pulling out the Lat and Lons I need from isoP. I will about a file for isoP request it will have Lat and Lons in it. I will then use this file to request the data I need from isoP.

```python
# Libraries
import numpy as np
import pandas as pd
```

```python
# Extract the station locations from the original data
isoData = pd.read_csv("Isoscape_Data.csv")
stations = isoData[['Station', 'Lat', 'Long']].drop_duplicates()
stationList = stations['Station'].tolist()
stations.set_index('Station', inplace=True)
stations
```

```python
# Now to open results test file
results = pd.read_csv("results_test.csv")
results
```

```python
# Attach the station locations to the results via lat and long
for stat in stationList:
    lat = stations.loc[stat, 'Lat']
    long = stations.loc[stat, 'Long']

    # Now find every instance of results where lat and long match and attach the station name
    results.loc[(results['Lat'] == lat) & (results['Long'] == long), 'Station'] = stat
```

```python
# Now save only the unique lat and longs and their corresponding station names
results_unique = results[['Lat', 'Long', 'Station']].drop_duplicates()
results_unique.to_csv("isoP_Request.csv", index=False)
```

```python
# Print earliest year and day of year as well as latest year and day of year
print("Earliest Year: ", results['Year'].min())
print("Earliest Day of Year: ", results['Day'].min())
print("Latest Year: ", results['Year'].max())
print("Latest Day of Year: ", results['Day'].max())
```
