# Results from experimental results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in the data
df = pd.read_csv('results.csv')
df.set_index('Date', inplace=True)
df
```

```python
# Plotting measured vallues on x-axis and predicted values on y-axis
# Draw line trhogh y = x all the way though the graph
# Calculate and display R2 value on the graph
# Calculate and display RMSE value on the graph
# Calculate and display MAE value on the graph

# Calculate R2
y_bar = df['O18Avg'].mean()
y_pred_bar = df['y_pred'].mean()
SSres = np.sum((df['O18Avg'] - df['y_pred'])**2)
SStot = np.sum((df['O18Avg'] - y_bar)**2)
R2 = 1 - SSres/SStot


plt.figure(figsize=(10, 10))
plt.scatter(df['O18Avg'], df['y_pred'], alpha=0.5, c='b')
plt.plot([-40,0], [-40, 0], c='black')
plt. text(-35, -5, 'R2 = ' + str(round(R2, 2)))
plt.xlabel('Measured O18')
plt.ylabel('Predicted O18')
plt.title('Measured vs Predicted O18')
plt.show()
```

```python
# Plotting y_pred and O18Avg over time of only the first 100 rows
plt.plot(df.index[250:500], df['y_pred'][250:500], label='Predicted')
plt.plot(df.index[250:500], df['O18Avg'][250:500], label='Actual')
plt.legend()
plt.title('Predicted vs Actual')
plt.xlabel('Date')
plt.ylabel('O18')
plt.show()
plt.close()
```
